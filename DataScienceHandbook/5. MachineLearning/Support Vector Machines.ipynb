```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns; sns.set()
from sklearn.datasets.samples_generator import make_blobs
```

```python
#Generate a example data
X, y = make_blobs(n_samples=50,centers=2, cluster_std=0.6,  random_state=0)
plt.scatter(X[:,0],X[:,1],c=y,cmap='autumn',s=50)
```

```python
X, y = make_blobs(n_samples=50,centers=2, cluster_std=0.6,  random_state=0)
plt.scatter(X[:,0],X[:,1],c=y,cmap='autumn',s=50,)
#draw separate lines
xfit = np.linspace(-1,3.5)
for m,b in [(1,0.8),(0.5,2),(-0.1,2.6)]:
    yfit = m*xfit + b
    plt.plot(xfit,yfit,color="Black")
    plt.fill_between(xfit,yfit-0.2,yfit+0.2,color="gray",alpha=0.3)
    
plt.xlim(-1,3.5)
plt.ylim(-1,6)
```
